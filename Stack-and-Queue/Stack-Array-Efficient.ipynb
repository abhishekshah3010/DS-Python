```python
# In this implementation we are starting with an already initialized list
class EmptyStackError(Exception):
    pass

class StackFullError(Exception):
    pass

class Stack:
    
    def __init__(self, max_size = 10):
        self.items = [None] * max_size 
        self.count = 0 # count will keep track of no of elements present in stack 
        
    def is_empty(self):
        return self.count == 0
    
    def size(self):
        return self.count # returns the number of elements in stack

    def is_full(self):
        return self.count == len(self.items)
    
    def push(self, x):
        if self.is_full():
            raise StackFullError("Stack is Full, Can't push")
        self.items[self.count] = x # new element is placed at the index count of the list
        self.count += 1 # the count is then incremented by 1
    
    def pop(self): # pop removes last element from the end of the list
        if self.is_empty():
            raise EmptyStackError("Stack is empty")
        x = self.items[self.count-1]
        self.items[self.count-1] = None
        self.count -= 1
        return x
        
    def peek(self): # returns the last element of the lists
        if self.is_empty():
            raise EmptyStackError("Stack is empty")
        return self.items[self.count-1] 
    
    def display(self):
        print(self.items)
```

```python
if __name__ == "__main__":
    st = Stack(8)
    
    while True:
        print("1.Push")
        print("2.Pop")
        print("3.Peek")
        print("4.Size")
        print("5.Display")
        print("6.Quit")
        
        choice = int(input("Enter your choice: "))
        
        if choice == 1:
            x = int(input("Enter the elements to be pushed: "))
            st.push(x)
        elif choice == 2:
            x = st.pop()
            print("Popped element is: ", x)
        elif choice == 3:
            print("Element at top is: ", st.peek())
        elif choice == 4:
            print("Size of stack is: ", st.size())
        elif choice == 5:
            st.display()
        elif choice == 6:
            break
        else:
            print("Wrong choice")
        print()
```
