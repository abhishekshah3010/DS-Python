```python
class EmptyQueueError(Exception):
    pass

class Queue:
    
    def __init__(self):
        self.items = []
        self.front = 0 # count will keep track of element at the front of the queue 
        
    def is_empty(self):
        return self.front == len(self.items)
    
    def size(self):
        return len(self.items) - self.front # returns the number of elements in queue
    
    def enqueue(self, item):
        self.items.append(item) # new element is placed at the index count of the list
        
    def dequeue(self): # pop removes last element from the end of the list
        if self.is_empty():
            raise EmptyQueueError("Queue is empty")
        
        x = self.items[self.front-1]
        self.items[self.front] = None
        self.front += 1
        return x
        
    def peek(self): # returns the last element of the lists
        if self.is_empty():
            raise EmptyQueueError("Queue is empty")
        return self.items[self.front] 
    
    def display(self):
        print(self.items)
```

```python
if __name__ == "__main__":
    qu = Queue()
    
    while True:
        print("1.Push")
        print("2.Pop")
        print("3.Peek")
        print("4.Size")
        print("5.Display")
        print("6.Quit")
        
        choice = int(input("Enter your choice: "))
        
        if choice == 1:
            x = int(input("Enter the elements to be pushed: "))
            qu.enqueue(x)
        elif choice == 2:
            x = qu.dequeue()
            print("Popped element is: ", x)
        elif choice == 3:
            print("Element at top is: ", qu.peek())
        elif choice == 4:
            print("Size of stack is: ", qu.size())
        elif choice == 5:
            qu.display()
        elif choice == 6:
            break
        else:
            print("Wrong choice")
        print()
```
